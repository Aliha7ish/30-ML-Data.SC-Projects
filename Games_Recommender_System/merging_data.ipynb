## Getting the best match between game names

```python
def get_best_match(query, choices, threshold=85, verbose=True):
    """
    Find the best match from a list of choices using fuzzy matching.
    If multiple matches have the same high score, return the one that matches the query exactly if present.
    """
    matches = process.extract(query, choices, scorer=fuzz.token_set_ratio, limit=3)

    # Filter matches above the threshold
    high_score_matches = [match for match in matches if match[1] >= threshold]

    # Prioritize exact match if present
    for match in high_score_matches:
        if match[0].lower() == query.lower():
            if verbose:
                print(f"Query = {query}, Matched game = {match[0]}")
                print(f"------------------ Exact Match -------------------")
            return match[0]

    # Return best fuzzy match
    if high_score_matches:
        best_match = high_score_matches[0]
        if verbose:
            print(f"Query = {query}, Matched game = {best_match[0]}")
            print(f"------------------ Fuzzy Match -------------------")
        return best_match[0]

    if verbose:
        print(f"Query = {query}, No match found above threshold.")
        
    return None
```

```python
unique_game_names = users['game_name_mod'].dropna().unique()
```

```python
tqdm.pandas()
games_list = games["name_mod"].tolist()
# Match each unique game name only once
match_dict = {
    name: get_best_match(name, games_list)
    for name in tqdm(unique_game_names)
}

users['matched_game'] = users['game_name_mod'].map(match_dict)
```

## Merge on matched games

```python
# load saved df
users = pd.read_csv("../data/users_matched_game_names.csv")
```

```python
users.columns
```

```python
users.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
users['matched_game'].isnull().sum()
```

```python
users[users['matched_game'].isnull()]
```

```python
# merge 2 dataframes based on matched game names
genre_cols = games.filter(like='genres_').columns.tolist()
tag_cols = games.filter(like='tags_').columns.tolist()

# Columns to merge: ID, name, and one-hot encoded metadata
columns_to_merge = ['game_id', 'name_mod', 'rating', 'metacritic'] + genre_cols + tag_cols

merged_users = users.merge(games[columns_to_merge], left_on='matched_game', right_on='name_mod', how='inner')
merged_users.head()
```
